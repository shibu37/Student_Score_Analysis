# Deleting columns which are not relevant to our analysis


```python
df=df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","05-10")
df.head()
```

# Gender distribution

```python
plt.figure(figsize =(5,5))
ax=sns.countplot(data = df , x= "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

#from the above chart we have analysed that the number of females in the data is more than the number of males 

```python
gb= df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot = True)
plt.title("Relation B/W parent's education and student's score")
plt.show()
```

```python
# from the above chart we have concluded that the education of the parents have  a good impact on their scores 
```

```python
gb1= df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot = True)
plt.title("Relation B/W parent's Marital Status & student's score")
plt.show()
```

```python
#so from above chart we can conclude that parent's marital status has negligible impact on a student's score
```

# Distribution of Ethnic Groups 


```python
groupA= df.loc[(df['EthnicGroup']== "group A")].count()
groupB= df.loc[(df['EthnicGroup']== "group B")].count()
groupC= df.loc[(df['EthnicGroup']== "group C")].count()
groupD= df.loc[(df['EthnicGroup']== "group D")].count()
groupE= df.loc[(df['EthnicGroup']== "group E")].count()

l= ["group A","group B","group C","group D","group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

plt.pie(mlist, labels=l , autopct= "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

# Relationship between Weekly Study Hours and Scores

```python

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Boxplot for Math Score vs Weekly Study Hours
sns.boxplot(x='WklyStudyHours', y='MathScore', data=df, ax=axes[0])
axes[0].set_title('Weekly Study Hours vs Math Score')

# Boxplot for Reading Score vs Weekly Study Hours
sns.boxplot(x='WklyStudyHours', y='ReadingScore', data=df, ax=axes[1])
axes[1].set_title('Weekly Study Hours vs Reading Score')

# Boxplot for Writing Score vs Weekly Study Hours
sns.boxplot(x='WklyStudyHours', y='WritingScore', data=df, ax=axes[2])
axes[2].set_title('Weekly Study Hours vs Writing Score')

# Show the plot
plt.tight_layout()
plt.show()
```

# Relationship B/W Test Preparation, Practice of Sport, and Scores

```python
# Test Preparation and Sports Practice vs Math Score
plt.figure(figsize=(10, 6))
sns.catplot(x='TestPrep', y='MathScore', hue='PracticeSport', kind='bar', data=df)
plt.title('Math Score vs Test Preparation and Sports Practice')
plt.show()

# Test Preparation and Sports Practice vs Reading Score
plt.figure(figsize=(10, 6))
sns.catplot(x='TestPrep', y='ReadingScore', hue='PracticeSport', kind='bar', data=df)
plt.title('Reading Score vs Test Preparation and Sports Practice')
plt.show()

# Test Preparation and Sports Practice vs Writing Score
plt.figure(figsize=(10, 6))
sns.catplot(x='TestPrep', y='WritingScore', hue='PracticeSport', kind='bar', data=df)
plt.title('Writing Score vs Test Preparation and Sports Practice')
plt.show()
```
